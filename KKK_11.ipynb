```python
Anushka_Present = len(data[data.Anushka_Present == 1])
Abhinav_Present = len(data[data.Abhinav_Present == 1])
Aastha_Present = len(data[data.Aastha_Present == 1])
Divyanka_Present   = len(data[data.Divyanka_Present== 1])
Mahek_Present       = len(data[data.Mahek_Present== 1])
Rahul_Present       = len(data[data.Rahul_Present== 1])
Sana_Present        = len(data[data.Sana_Present== 1])
Saurabh_Present     = len(data[data.Saurabh_Present== 1])
Shweta_Present      = len(data[data.Shweta_Present== 1])
Varun_Present       = len(data[data.Varun_Present== 1])
Vishal_Present      = len(data[data.Vishal_Present== 1])
NIkki_Present = len(data[data.NIkki_Present== 1])
```

```python
fig = plt.figure(figsize = (10, 5)) #lxb image
present = [Abhinav_Present  , 
Anushka_Present,     
Aastha_Present,  
Divyanka_Present,    
Mahek_Present,       
Rahul_Present,       
Sana_Present,       
Saurabh_Present,     
Shweta_Present,      
Varun_Present,       
Vishal_Present,      
NIkki_Present,       
        ]
names= ["Abhinav" , 
"Anushka",     
"Aastha",  
"Divyanka",    
"Mahek",       
"Rahul",       
"Sana",       
"Saurabh",     
"Shweta",      
"Varun",       
"Vishal",      
"NIkki"
        ]   
plt.bar(names, present, color ='red',width = 0.4)
plt.xlabel("Candidate Name") #title of x axis
plt.ylabel("Candidate Present") #title of y axis
plt.title("Candidate Present vs Name") #title of the graph
plt.show() #this excecutes print statement for graph
```

```python
Abhinav_Out         =  len(data[data.Abhinav_Out==0])
Anushka_Out         = len(data[data.Anushka_Out==0])
Aastha_Out          = len(data[data.Aastha_Out==0])
Arjun_Out           =len(data[data.Arjun_Out==0])
Divyanka_Out        =len(data[data.Divyanka_Out==0])
Mahek_Out           =len(data[data.Mahek_Out==0])
Sana_Out            =len(data[data.Sana_Out==0])
Rahul_Out           =len(data[data.Rahul_Out==0])
Saurabh_Out         =len(data[data.Saurabh_Out==0])
Shweta_Out          =len(data[data.Shweta_Out==0])
Varun_Out           =len(data[data.Varun_Out==0])
Vishal_Out          =len(data[data.Vishal_Out==0])
NIkki_Out        =len(data[data.NIkki_Out==0])
```

```python
Abhinav_Won         =  len(data[data.Abhinav_Won_or_Did_not_play==0])
Anushka_Won       = len(data[data.Anushka_Won_or_Did_not_play==0])
Aastha_Won          = len(data[data.Aastha_Won_or_Did_not_play==0])
Arjun_Won           =len(data[data.Arjun_Won_or_Did_not_play==0])
Divyanka_Won        =len(data[data.Divyanka_Won_or_Did_not_play==0])
Mahek_Won           =len(data[data.Mahek_Won_or_Did_not_play==0])
Sana_Won            =len(data[data.Sana_Won_or_Did_not_play==0])
Rahul_Won           =len(data[data.Rahul_Won_or_Did_not_play==0])
Saurabh_Won         =len(data[data.Saurabh_Won_or_Did_not_play==0])
Shweta_Won          =len(data[data.Shweta_Won_or_Did_not_play==0])
Varun_Won           =len(data[data.Varun_Won_or_Did_not_play==0])
Vishal_Won          =len(data[data.Vishal_Won_or_Did_not_play==0])
NIkki_Won       =len(data[data.NIkki_Won_or_Did_not_play==0])
```

```python
Abhinav_Eli_Won_or_Did_not_play     = (Abhinav_Won / Abhinav_Out)*100
Anushka_Eli_Won_or_Did_not_play     = (Anushka_Won/Anushka_Out)*100
Aastha_Eli_Won_or_Did_not_play  =   (Aastha_Won/Aastha_Out)*100
Arjun_Eli_Won_or_Did_not_play    =   (Arjun_Won/Arjun_Out)*100
Divyanka_Eli_Won_or_Did_not_play  =  (Divyanka_Won/Divyanka_Out)*100
Mahek_Eli_Won_or_Did_not_play    =   (Mahek_Won/Mahek_Out)*100
Rahul_Eli_Won_or_Did_not_play      = (Rahul_Won/Rahul_Out)*100
Sana_Eli_Won_or_Did_not_play      =  (Sana_Won/Sana_Out)*100
Saurabh_Eli_Won_or_Did_not_play =    (Saurabh_Won/Saurabh_Out)*100
Shweta_Eli_Won_or_Did_not_play =     (Shweta_Won/Shweta_Out)*100
Varun_Eli_Won_or_Did_not_play   =   (Varun_Won/Varun_Out)*100
Vishal_Eli_Won_or_Did_not_play   =   (Vishal_Won/Vishal_Out)*100
NIkki_Eli_Won_or_Did_not_play= (NIkki_Won/NIkki_Out)*100
```

```python
fig = plt.figure(figsize = (10, 5)) #lxb image

elimination = [Abhinav_Eli_Won_or_Did_not_play,
Anushka_Eli_Won_or_Did_not_play,   
Aastha_Eli_Won_or_Did_not_play,     
Mahek_Eli_Won_or_Did_not_play,     
Rahul_Eli_Won_or_Did_not_play,       
Sana_Eli_Won_or_Did_not_play,        
Saurabh_Eli_Won_or_Did_not_play,     
Shweta_Eli_Won_or_Did_not_play,      
Varun_Eli_Won_or_Did_not_play,      
Vishal_Eli_Won_or_Did_not_play,      
NIkki_Eli_Won_or_Did_not_play,
      ]

names= ["Abhinav" , 
"Anushka",     
"Aastha",  
"Mahek",       
"Rahul",       
"Sana",       
"Saurabh",     
"Shweta",      
"Varun",       
"Vishal",      
"NIkki",
        ]   
plt.bar(names,elimination, color ='red',width = 0.4)
plt.xlabel("Candidate Name") #title of x axis
plt.ylabel("Candidate Eliminated") #title of y axis
plt.title("Candidate eliminated vs Name") #title of the graph
plt.show() #this excecutes print statement for graph
```

```python
plt.pie(elimination, labels = names, radius = 2.0, autopct = '%0.1f%%', ) 
plt.suptitle('Percentage of candidates invested', x = 0.25, 
y = 1.5, color = 'black') 
plt.show()
```
